```python
import pandas as pd
import json
```

```python



# Read files into a dataframe
with open('data_json.txt', 'r') as fp:
    data = json.load(fp)

pd.DataFrame.from_dict({(i,j): data[i][j] 
                           for i in data.keys() 
                           for j in data[i].keys()},
                       orient='index')



```
